# K-mean Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans2 

a = np.random.multivariate_normal([0, 6], [[2, 1], [1, 1.5]], size=200)
b = np.random.multivariate_normal([2, 0], [[1, -1], [-1, 3]], size=400)
c = np.random.multivariate_normal([6, 4], [[5, 0], [0, 1.2]], size=250)
z = np.concatenate((a, b, c))

np.random.shuffle(z)

plt.scatter(z[:, 0], z[:, 1])
plt.show()

print(z)
```

```python
k = 3

centroids, labels = kmeans2(z, k, minit="points")
print("Centroids: ", centroids)
print("Labels: ", labels)
```

```python
cluster_1 = z[labels == 0]
cluster_2 = z[labels == 1]
cluster_3 = z[labels == 2]
print("Cluster 1 values: ", cluster_1)
print("Cluster 2 values: ", cluster_2)
print("Cluster 3 values: ", cluster_3)
```

```python
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], label="Cluster_1")
plt.scatter(cluster_2[:, 0], cluster_2[:, 1], label="Cluster_2")
plt.scatter(cluster_3[:, 0], cluster_3[:, 1], label="Cluster_3")
plt.scatter(centroids[:, 0], centroids[:, 1], label="Centroids", color="#000000")
plt.show()
```

# Interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(11)
Y = np.array([2.0, 1.9, 1.7, 1.5, 0.5, 0.0, 0.8, 2.0, 0.9, 0.2, 2.0])

plt.plot(X, Y, "o:")
plt.show()
```

```python
from scipy.interpolate import interp1d

# Interpolation is finding a point between two points

predict = interp1d(X, Y, kind="quadratic") # returns a function

X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print("X: 1.5, Y :",predict(1.5))
```

```python
predict = interp1d(X, Y, kind="linear") # returns a function

X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print("X: 1.5, Y :",predict(1.5))
```

```python
predict = interp1d(X, Y, kind="nearest") # returns a function

X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print("X: 1.5, Y :",predict(1.5))
```

```python
predict = interp1d(X, Y, kind="zero") # returns a function

X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print("X: 1.5, Y :",predict(1.5))
```

```python
predict = interp1d(X, Y, kind="cubic") # returns a function

X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print("X: 1.5, Y :",predict(1.5))
```

# Linear Algebra

```python
import numpy as np
import scipy.linalg as linalg

a = np.array([[3, 4] , [2, 5]])  # coefficients
b = np.array([10 , 9])  # constants

x_val, y_val = linalg.solve(a , b)
print("Solution: ", linalg.solve(a , b))

print("x value: ", x_val)
print("y value: ", y_val)
```

```python
array = np.array([[1, 2, 3] , [4, 5, 6], [8, 8, 9]])
print(array)
```

```python
print("Determinent: ", linalg.det(array))
```

```python
print("Inverse: \n", linalg.inv(array))
```

```python
print("Matrix - Inverse Matrix \n", array @ linalg.inv(array) )
```

```python
array2 = np.array([[1, -2, 0] , [-2, 1, 0], [0, 0, 3]])
print(array2)
```

```python
# Eigen Value and Vector
# vector that undergoes pure scaling without rotation
eg_value, eg_vector = linalg.eig(array2)

print("Eigen Value: \n", eg_value)
print("Eigen Vector: \n", eg_vector)
```

```python
# Singular Value Decomposition

P = eg_vector  # Matrix of eigenvectors
P_inv = linalg.inv(eg_vector)  # Inverse of the matrix of eigenvectors

D = []  # Initialize an empty list to store rows

for index, value in enumerate(eg_value):  # Loop through eigenvalues with their indices
    row = np.zeros((3))  # Create a row of zeros with 3 elements
    row[index] = value  # Assign the eigenvalue to the correct position in the row
    D.append(row)  # Append the row to the list D

D = np.array(D)  # Convert the list D to a NumPy array
print("Diagonal Matrix D: \n", D) 
```

```python
print("Return Original Matrix: \n", P @ D @ P_inv)
```

# Integration

```python
import math
from scipy.integrate import quad

def f(x: float) -> float:
    return math.sin(x)

result = quad(f, 0, math.pi/2 ) # function, lower_bound, upper_bound
print(result)
# returns answer plus error
```

```python
import math
from scipy.integrate import dblquad

def f(x, y):
    return x ** 2 + y ** 2

area, error = dblquad(f, -5, 5, -10, 10)
#  function, lower_bound and upper_bound of x, lower_bound and upper_bound of y
print(area , error)
```

```python
import math
from scipy.integrate import nquad

def f(x, y, z):
    return x * y * z

area, error = nquad(f, [[0, 2], [-2, 0], [1, 2]])
print(area ,"  ", error)
```

# Optimization

```python
import numpy as np
from scipy.optimize import minimize

# Minimize a**2 + b**2  Initial Condition: a= 200 and b= 500
def objective(x):
    a = x[0]
    b = x[1]
    return (a - 4) ** 2 + (b - 3) ** 2

x0 = np.array([200, 500]) # Initial Condition
solution = minimize(objective, x0, method="SLSQP")
print(solution)
```

```python
def objective(x):
    a = x[0]
    b = x[1]
    return (a - 4) ** 2 + (b - 3) ** 2

x0 = np.array([200, 500]) # Initial Condition
solution = minimize(objective, x0, method="SLSQP")
print(solution)
```

```python
# Minimize a**2 + b**2  Initial Condition: a= 700 and b= -600
# Constraint: a + b = 100
# Constraint: ab > 100

def constraint1(x):
    a = x[0]
    b = x[1]
    return a + b - 100

def constraint2(x):
    a = x[0]
    b = x[1]
    return a * b - 100

cons = [{
    "type": "eq",
    "fun": constraint1
}, {
    "type": "ineq",
    "fun": constraint2
}]

solution = minimize(objective, x0, method="SLSQP", constraints=cons)
print(solution)
```

```python
# bounds = 10 < a,b < 50

bound= np.array([10 , 50])
bounds = (bound, bound)

solution = minimize(objective, x0, method="SLSQP", bounds=bounds)
print(solution)
```
